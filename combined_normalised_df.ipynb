```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
%matplotlib inline
```

```
df1 = pd.read_csv('/content/Normalised_Merge_df-2.csv',index_col=False)
df1 = df1[df1.columns[1:]]
df1.columns = df1.columns.str.replace(' ', '')
df1['station'] = 'Cross Barr'

df2 = pd.read_csv('/content/Normalised_Goliad_05.csv',index_col=False)
df2 = df2[df2.columns[1:]]
df2['station'] = 'Goliad'

df3 = pd.read_csv('//content/Normalised_Hobbit_05.csv',index_col=False)
df3 = df3[df3.columns[1:]]
df3 = df3.replace(np.nan, 0)
df3['station'] = 'Hobbit'

df4 = pd.read_csv('/content/Normalised_Katy_05.csv',index_col=False)
df4 = df4[df4.columns[1:]]
df4['station'] = 'Katy'


df5 = pd.read_csv('/content/Normalised_Lighting_05.csv',index_col=False)
df5 = df5[df5.columns[1:]]
df5['station'] = 'Lightning'


df6 = pd.read_csv('/content/Normalised_Reagon_05.csv',index_col=False)
df6 = df6[df6.columns[1:]]
df6['station'] = 'Reagon'

df7 = pd.read_csv('/content/Normalised_Turkey_05.csv',index_col=False)
df7 = df7[df7.columns[1:]]
df7['station'] = 'Reagon'


#cross_bar_df3 = pd.read_csv('/content/Normalised_CroosBar_05.csv',index_col=False)
#cross_bar_df3 = cross_bar_df3[cross_bar_df3.columns[1:13]]
```
